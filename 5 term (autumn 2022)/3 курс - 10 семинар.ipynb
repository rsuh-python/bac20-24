1. Нужно открыть текстовый файл и считать из него текст в виде списка слов без пунктуации.

```python
text = ''
with open('/content/1.txt', 'r') as file:
  for line in file:
    for char in line:
      if char.isalpha() or char == ' ':
        text += char
text
```
